# Exp 4 : Implementing XOR using backpropagation

```python
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return x*(1-x)
```

```python
inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
expected_output=np.array([[0],[1],[1],[0]])
```

```python
epochs = 10000
lr=0.1
#Squared error + plot wrt iterations
inputLayerNeurons, hiddenLayerNeurons, outputLayerNeurons = 2,2,1
```

```python
# Random weights and bias initialization
hidden_weights = np.random.uniform(size=(inputLayerNeurons,hiddenLayerNeurons))
hidden_bias = np.random.uniform(size=(1, hiddenLayerNeurons))
output_weights = np.random.uniform(size=(hiddenLayerNeurons, outputLayerNeurons))
output_bias = np.random.uniform(size=(1, outputLayerNeurons))
predicted_output=[]
```

```python
print("Initial hidden weights: \n",hidden_weights)
print("Initial hidden biases: \n", hidden_bias)
print("Initial output weights: \n",output_weights)
print("Initial output bias: \n",output_bias)
```

## Training algorithm

```python
er=[]
```

```python
for _ in range(epochs):
    #FORWARD PROPAGATION
    hidden_layer_activation = np.dot(inputs, hidden_weights)
    hidden_layer_activation+=hidden_bias
    hidden_layer_output = sigmoid(hidden_layer_activation)
    
    output_layer_activation = np.dot(hidden_layer_output, output_weights)
    output_layer_activation+=output_bias
    predicted_output = sigmoid(output_layer_activation)
    
    #BACKPROPAGATION
    error = expected_output - predicted_output
    er.append(sum(error**2)/len(error))
    d_predicted_output = error*sigmoid_derivative(predicted_output)
    
    error_hidden_layer = d_predicted_output.dot(output_weights.T)
    d_hidden_layer = error_hidden_layer*sigmoid_derivative(hidden_layer_output)
    
    #UPDATING WEIGHTS AND BIASES
    output_weights+=hidden_layer_output.T.dot(d_predicted_output)*lr
    output_bias+=np.sum(d_predicted_output, axis=0, keepdims=True)*lr
    hidden_weights+=inputs.T.dot(d_hidden_layer)*lr
    hidden_bias+=np.sum(d_hidden_layer, axis=0, keepdims=True)*lr
    print(_)
    print(error)
```

## Error Plot

```python
from matplotlib import pyplot as plt
plt.plot([i for i in range(10000)],er)
plt.title('Squared Error vs Epochs')
plt.xlabel('Squared Error')
plt.ylabel('Epochs')
```

```python
print("Final hidden weights: \n",hidden_weights)
print("Final hidden biases: \n", hidden_bias)
print("Final output weights: \n",output_weights)
print("Final output bias: \n",output_bias)

print("Output from neural network after 10000 epochs: ", predicted_output)
```

# Implementing on Iris dataset

```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
df1=df[df.target==1] #versicolor
df2=df[df.target==0] #setosa
```

```python
# versicolor's head
df1.head()
```

```python
inputs = df2[['sepal length (cm)','petal length (cm)']].append(df1[['sepal length (cm)','petal length (cm)']], ignore_index=True).to_numpy()
expected_output=df2.target.append(df1.target, ignore_index=True).to_numpy()
expected_output=expected_output.reshape(100,1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, expected_output, test_size=1)
```

```python
er=[]
for _ in range(epochs):
    #FORWARD PROPAGATION
    hidden_layer_activation = np.dot(x_train, hidden_weights)
    hidden_layer_activation+=hidden_bias
    hidden_layer_output = sigmoid(hidden_layer_activation)
    
    output_layer_activation = np.dot(hidden_layer_output, output_weights)
    output_layer_activation+=output_bias
    predicted_output = sigmoid(output_layer_activation)
    
    #BACKPROPAGATION
    error = y_train-predicted_output
    er.append(sum(error**2)/len(error))
    d_predicted_output = error*sigmoid_derivative(predicted_output)
    
    error_hidden_layer = d_predicted_output.dot(output_weights.T)
    d_hidden_layer = error_hidden_layer*sigmoid_derivative(hidden_layer_output)
    
    #UPDATING WEIGHTS AND BIASES
    output_weights+=hidden_layer_output.T.dot(d_predicted_output)*lr
    output_bias+=np.sum(d_predicted_output, axis=0, keepdims=True)*lr
    hidden_weights+=x_train.T.dot(d_hidden_layer)*lr
    hidden_bias+=np.sum(d_hidden_layer, axis=0, keepdims=True)*lr
```

```python
from matplotlib import pyplot as plt
plt.plot([i for i in range(epochs)],er)
plt.title('Squared Error vs Epochs')
plt.xlabel('Squared Error')
plt.ylabel('Epochs')
```

```python
def classifier_(x):
    hidden_layer_activation = np.dot(x, hidden_weights)
    hidden_layer_activation+=hidden_bias
    hidden_layer_output = sigmoid(hidden_layer_activation)
    
    output_layer_activation = np.dot(hidden_layer_output, output_weights)
    output_layer_activation+=output_bias
    predicted_output = sigmoid(output_layer_activation)
    return predicted_output
```
